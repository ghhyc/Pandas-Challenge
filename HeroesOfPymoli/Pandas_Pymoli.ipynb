```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
df = pd.read_csv(file_to_load)
df.head()


```

```python
#compute player

player_demo = df.loc[:, ["SN","Age","Gender"]]
player_demo = player_demo.drop_duplicates()
# [0] drop index
tot_players = player_demo.count()[0] 
pd.DataFrame ({"Total Players": [tot_players]})
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_item_ct = len(df['Item ID'].unique())
ave_price_items = df["Price"].mean()
purchase_count = len(df['Price'])
price_sum = df['Price'].sum()
price_sum

sum_df = pd.DataFrame({'Number of Unique item': [unique_item_ct],
                       'Ave Purchase Price': [ave_price_items],
                       'Total No of Purchases': [purchase_count],
                       'Total Revenue': [price_sum]
                       })
sum_df['Ave Purchase Price'] = sum_df['Ave Purchase Price'].map('${:.2f}'.format)
sum_df['Total No of Purchases'] = sum_df['Total No of Purchases'].map('${:.2f}'.format)
sum_df['Total Revenue'] = sum_df['Total Revenue'].map('${:.2f}'.format)

sum_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 150]
bin_labels = ['<10', '10-14', '15-19', '20-24','25-29','30-34', '35-39','40+']

df['Total Count'] = pd.cut(df['Age'], bins, labels=bin_labels)
df_age1 = df[['Total Count','SN']].drop_duplicates(subset = 'SN')

age_dem_sum = df_age1.groupby('Total Count').count()
age_counts = age_dem_sum['SN']

age_dem_percent = age_counts / 576

age_dem_df = pd.DataFrame({
        'Total Count':age_counts,
        'Percentage of Players':age_dem_percent
})
age_dem_df['Percentage of Players'] = age_dem_df['Percentage of Players'].map('{:.2%}'.format)

age_dem_df
```

## Top Spenders

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
df_sn_3 = df.groupby(['Item ID','Item Name'])

purchase_cnt3 = df_sn_3['Gender'].count()

ave_spend3 = round(df_sn_3['Price'].mean(),2)

tot_purchase_3 = round(df_sn_3['Price'].sum(),2)

popular_items = pd.DataFrame({
            'Purchase Count':purchase_cnt3,
            'Average Purchase Price':ave_spend3,
            'Total Purchase Value':tot_purchase_3
})
#popular_items.sort_values(by=['Purchase Count'],ascending=False)
popular_items = popular_items.sort_values('Purchase Count',ascending=False)

popular_items['Average Purchase Price'] = popular_items['Average Purchase Price'].map('${:.2f}'.format)

popular_items['Total Purchase Value'] = popular_items['Total Purchase Value'].map('${:.2f}'.format)
popular_items


```
